# Handling Missing Values

```python
dftrain.isna().sum()
```

```python
dftrain.describe()
```

```python
dftest.isna().sum()
```

```python
#Handling missing values in train dataset
col_numeric = dftrain.select_dtypes(include=(['int64', 'float64'])).columns
for col in col_numeric:
    dftrain[col].fillna(dftrain[col].median(), inplace=True)

col_obj = dftrain.select_dtypes(include=(['object'])).columns
for col in col_obj:
    dftrain[col].fillna(dftrain[col].mode().values[0], inplace=True)

dftrain.isna().sum()
```

```python
#Handling missing values in test dataset
col_numeric = dftest.select_dtypes(include=(['int64', 'float64'])).columns
for col in col_numeric:
    dftest[col].fillna(dftest[col].median(), inplace=True)

col_obj = dftest.select_dtypes(include=(['object'])).columns
for col in col_obj:
    dftest[col].fillna(dftest[col].mode().values[0], inplace=True)

dftest.isna().sum()
```

# Handling Outlier & Skewness

```python
dftrain.plot(kind = 'box', subplots = True, layout = (3,9),sharex = False, sharey = False, 
        figsize = (20,30), color ="#ff2655")
```

```python
dftest.plot(kind = 'box', subplots = True, layout = (3,9),sharex = False, sharey = False, 
        figsize = (20,30), color ="#ff2655")
```

```python
cols = ['age','no_of_children']
dftrain[cols].hist(figsize=(20,20))
```

```python
dftest[cols].hist(figsize=(20,20))
```

```python
#Log transformation to remove skewness
for col in cols:
    dftrain[col] = dftrain[col].apply(lambda x: np.log(x + 1))
    dftest[col] = dftest[col].apply(lambda x: np.log(x + 1))
```

```python
dftrain[cols].hist(figsize=(10,5))
```

```python
dftest[cols].hist(figsize=(10,5))
```

```python
#Handling Outlier - Train
Q1 = dftrain[cols].quantile(0.25)
Q3 = dftrain[cols].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
dftrain[cols]=np.where(dftrain[cols]>upper_whisker,upper_whisker,np.where(dftrain[cols]<lower_whisker,lower_whisker,dftrain[cols]))
```

```python
#Handling Outlier - Test
Q1 = dftest[cols].quantile(0.25)
Q3 = dftest[cols].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
dftest[cols]=np.where(dftest[cols]>upper_whisker,upper_whisker,np.where(dftest[cols]<lower_whisker,lower_whisker,dftest[cols]))
```
